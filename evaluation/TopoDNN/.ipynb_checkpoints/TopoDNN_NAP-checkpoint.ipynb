```python
from keras.models import load_model
import pandas as pd
import pickle
import tensorflow as tf
import numpy as np
import keras
import matplotlib.pyplot as plt
import shap
import seaborn as sns
import os
```

```python
## List of models
print("\n".join([f for f in os.listdir("../../models/TopoDNN/topodnnmodels/")]))
```

```python
#Can change to '', '_pt0','_pt', '_30', or '_standardize_pt'
mode = ''
#Load model
model = load_model("../../models/TopoDNN/topodnnmodels/topodnnmodel"+mode)
model.summary()


#Loading data
df_test = pd.read_pickle("../../datasets/topoprocessed/test.pkl")
if mode == '_standardize_pt':
    pt_cols = [col for col in feat_names if 'pt' in col]
    df_test[pt_cols] = (df_test[pt_cols] - df_test[pt_cols].mean())/df_test[pt_cols].std()
feats = np.array(df_test.loc[:,df_test.columns != 'is_signal_new'])
means = np.array(df_test.loc[:,df_test.columns != 'is_signal_new'].mean(0))
labels = np.array(df_test.loc[:,df_test.columns == 'is_signal_new']).reshape(-1)
feat_names = list(df_test.columns)[:-1]

print("Features shape: ", feats.shape)
print("Means shape: ", means.shape)
print("Results shape: ", labels.shape)
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score, precision_recall_curve

def eval(model, mask_index = -1, mask_mode = 'mean'):
    data = np.copy(feats)
    if mask_index >= 0 and mask_index < len(feat_names):
        if mask_mode == 'mean':
            data[:, mask_index] = means[mask_index]
        else:
            data[:, mask_index] = 0.
    return model(data)

def eval2(model, mask_index = -1, mask_mode = 'mean'):
    data = np.copy(feats)
    if mask_index >= 0 and mask_index < len(feat_names):
        if mask_mode == 'mean':
            data[:, mask_index] = means[mask_index]
        else:
            data[:, mask_index] = 0.
    Nlayers = len(model.layers)
    for ii, layer in enumerate(model.layers):
        data = layer(data)
        if ii == Nlayers - 2:
            presig = np.array(data)
        if ii == Nlayers - 1:
            sig = np.array(data)
    return presig, sig

def dAUC_chart(dAUC_vals, tags, fname):
    dAUC_vals = dAUC_vals[::-1]
    tags = tags[::-1]
    pos = np.arange(len(dAUC_vals))
    if len(tags) < 10:
        plt.figure(figsize=(4,4))
    else:
        plt.figure(figsize=(10,10))
    plt.barh(pos, dAUC_vals, align='center')
    plt.yticks(pos, tags)
    plt.xlabel('Percent Drop in AUC')
    plt.tight_layout()
    #plt.savefig(fname)
    plt.show()
```

```python
baseline_presig, baseline_results = eval2(model)
auc_baseline = roc_auc_score(labels, baseline_results)*100
print("Baseline model's AUC: {:.4f}%".format(auc_baseline))
```

```python
tags = []
dAUC_vals = []
dresult = []
dresult_presig = []

for fid, feat in enumerate(feat_names):
    if feat == "phi_0" or feat == "eta_0":
        continue
    this_presig, this_results = eval2(model, mask_index = fid, mask_mode='zero')
    dresult_presig.append(((this_presig - baseline_presig)/baseline_presig).reshape(-1))
    dresult.append(((this_results - baseline_results).reshape(-1)))
    this_auc = roc_auc_score(labels, this_results)*100
    dAUC_vals.append(auc_baseline - this_auc)
    tags.append(feat)
```

```python
sorted_tags = []
sorted_vals = []
for val, tag in sorted(zip(dAUC_vals, tags), reverse = True)[:20]:
    sorted_tags.append(tag)
    sorted_vals.append(val)
    
dAUC_chart(sorted_vals, sorted_tags, fname='')
```

```python
dresult = np.transpose(np.array(dresult))
dresult_presig = np.transpose(np.array(dresult_presig))
```
